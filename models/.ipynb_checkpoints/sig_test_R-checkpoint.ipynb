```r
library(tidyverse)
library(ggplot2)
options(repr.plot.width=4, repr.plot.height=4)
df <- read_csv("../data/processed/cleaned_df.csv")
```

## 1. accept_marketing VS coming back and purchase

H0: Sample takers who accept marketing emails have the same rate of coming back and purchase of those who refuse marketing emails.

Ha: Sample takers who accept marketing emails have a higher rate of coming back and purchase of those who refuse marketing emails.

```r
buy_vs_marketing <- table(df$buy, df$accepts_marketing)
ggplot(df,aes(x = accepts_marketing, fill = buy)) + 
    geom_bar(position = "fill")
```

```r
chisq.test(buy_vs_marketing)
```

## 2. free_shipping VS coming back and purchase

H0: Sample takers who paid for their sample have the same rate of coming back and purchase of those who didn't.

Ha: Sample takers who paid for their sample have a higher rate of coming back and purchase of those who didn't.

```r
buy_vs_free <- table(df$buy, df$free_shipping)
ggplot(df,aes(x = free_shipping, fill = buy)) + 
    geom_bar(position = "fill")
```

```r
chisq.test(buy_vs_free)
```

## 3. location VS coming back and purchase

```r
buy_vs_location <- table(df$buy, df$location)
chisq.test(buy_vs_location)
```
