```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
pd.options.display.max_rows = 15
```

```python
df = pd.read_csv("../data/processed/cleaned_df.csv")
df.head()
```

```python
# shows the total number of customers from each location
df.groupby('location').count().sort_values(by = 'buy', ascending = False)[0:9][['buy']]
```

```python
print(f"The total number of customers in the top 4 locations with highest number of sample-takers is {df.groupby('location').count().sort_values(by = 'buy', ascending = False)[0:4]['buy'].sum()}.")
print(f"The total number of customers is {len(df)}.")
```

- the amount of customers in the top 4 locations are  each representative of the total number of customers overall, so we will use them for analysis

```python
top_4 = df[(df['location'].str.contains('ONTARIO, CANADA')) | (df['location'].str.contains('BRITISH COLUMBIA, CANADA')) | (df['location'].str.contains('ALBERTA, CANADA')) | (df['location'].str.contains('QUEBEC, CANADA'))]
```

```python
top_4.head()
```

### Ideal Customer Features

```python
for i in range(4):
    location = top_4.groupby(['accepts_marketing', 'free_shipping', 'location']
                            ).mean().sort_values(by = 'buy', ascending = False).index[i][2]
    print(f"The conversion rate of customers from {location} that accept marketing and don't have free shipping is {round(top_4.groupby(['accepts_marketing', 'free_shipping', 'location']).mean().sort_values(by = 'buy', ascending = False)['buy'].iloc[i], 3)}.")
```

```python
print(f"The overall conversion rate from top 4 locations is {round(top_4['buy'].mean(), 3)}.")
```

```python
# shows the total number of custoemrs in each category
top_4.groupby(['accepts_marketing', 'free_shipping', 'location']).count(
            ).sort_values(by = 'accepts_marketing', ascending = False
             ).sort_values(by = 'free_shipping', ascending = True
                          ).drop(columns = ['customer_id', 'ordered_month', 'gender', 'product_type',
                                           'skin_type', 'fv_site'])[0:4]
```

```python
no_ship_accept_mark = round(top_4.groupby(['accepts_marketing', 'free_shipping', 'location']).count(
            ).sort_values(by = 'accepts_marketing', ascending = False
             ).sort_values(by = 'free_shipping', ascending = True
                          )['buy'][0:4].sum()/len(top_4), 3)
print(f"The proportion of customers from top 4 locations that accept marketing and don't have free shipping is around {no_ship_accept_mark}.")
```

```python
# shows the conversion rate for each category
top_4.groupby(['accepts_marketing', 'free_shipping', 'location']).mean().sort_values(by = 'buy', ascending = False).drop(columns = ['customer_id', 'ordered_month'])[0:4]
```

```python
df = top_4.groupby(['accepts_marketing', 'free_shipping', 'location']).mean().sort_values(by = 'buy', ascending = False).drop(columns = ['customer_id', 'ordered_month'])[0:4]
best_conversions = list(df.groupby('location').mean().sort_values(by = 'buy', ascending = False)['buy'])
best_conversions.append(top_4['buy'].mean())
locations = ['BRITISH COLUMBIA, CANADA', 'ALBERTA, CANADA', 'ONTARIO, CANADA', 'QUEBEC, CANADA', 'Baseline']
yes_mark_no_ship_df = pd.DataFrame([locations, best_conversions]).T
yes_mark_no_ship_df.columns = ['Location', 'Conversion Rate']
# yes_mark_no_ship_df
```

```python
bar_plot = sns.catplot(x="Conversion Rate", y='Location', hue="Location", data=yes_mark_no_ship_df,
                       kind="bar", palette="muted")
bar_plot.fig.subplots_adjust(top=0.9)
bar_plot.fig.suptitle('Conversion Rate of Customers Who Accepted Marketing and Had No Free Shipping Compared to Total Baseline')
```

#### Conclusion:

In top 4 locations, especially British Columbia, customers that accept marketing and don't have free shipping display higher conversion rates. 

#### Limitation:

Only 1% customers from top 4 locations accept marketing and don't have free shipping.

### SHAP Analysis

```python
import shap
import pickle
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, LabelBinarizer
from sklearn.utils.validation import column_or_1d
from sklearn.pipeline import Pipeline, make_pipeline
```

```python
df = pd.read_csv('../data/processed/cleaned_df.csv')
X = df.drop(columns = ['customer_id', 'buy'])
y = df['buy']
seed = 123
# get test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=seed)
# get train/valid set
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=0.4, stratify=y_train, random_state=seed)
train_df = X_train.copy()
train_df['buy'] = y_train
valid_df = X_valid.copy()
valid_df['buy'] = y_valid
test_df = X_test.copy()
test_df['buy'] = y_test
```

```python
le = LabelEncoder()
```

```python
X_train_num = X_train.apply(LabelEncoder().fit_transform)
```

```python
X_train_num
```

```python
model = XGBClassifier(scale_pos_weight=(y_train == False).sum()/(y_train == True).sum())
```

```python
model.fit(X_train_num, y_train)
```

```python
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X_train_num)
```
