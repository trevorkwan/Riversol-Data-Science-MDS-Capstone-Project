# Imports

## Import libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score, cross_validate
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, LabelBinarizer
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.utils.validation import column_or_1d
from imblearn.over_sampling import SMOTENC
from imblearn.under_sampling import RandomUnderSampler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.compose import ColumnTransformer
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, matthews_corrcoef, roc_auc_score, plot_confusion_matrix
from sklearn.metrics import confusion_matrix, roc_curve, plot_confusion_matrix, classification_report
import time
import warnings
warnings.filterwarnings('ignore') 
plt.style.use('ggplot')
plt.rcParams.update({'font.size': 16,
                     'axes.labelweight': 'bold',
                     'figure.figsize': (8,6)})
```

## Define custom functions

```python
def run_models(model_dict, preprocessor, X_train, y_train, X_valid, y_valid):
    results_dict = {}
    fig, ax = plt.subplots(2, 3, figsize=(14, 9))
    for i, (model_name, model) in enumerate(models.items()):
        print(f"Running: {model_name}")
        t1 = time.time()
        clf = Pipeline(steps = [('preprocessor', preprocessor),
                                ('classifier', model)])
        clf.fit(X_train, y_train)
        p_train = clf.predict(X_train)
        p_valid = clf.predict(X_valid)
        results_dict[model_name] = return_metrics(y_train, p_train, y_valid, p_valid)
        plot_confusion(confusion_matrix(y_valid, p_valid), ax=ax[i // 3, i % 3], title=model_name)
        print(f"  Time taken: {time.time() - t1:.2f}s")
    plt.tight_layout()
    return (pd.DataFrame(results_dict,
                         index = ["train_accuracy", "valid_accuracy", "precision", "recall", "f1"])
              .transpose()
              .round(2)), fig

def return_metrics(y_train, y_train_predicted, y_valid, y_valid_predicted):
    train_acc = accuracy_score(y_train, y_train_predicted)
    test_acc = accuracy_score(y_valid, y_valid_predicted)
    p = precision_score(y_valid, y_valid_predicted)
    r = recall_score(y_valid, y_valid_predicted)
    f1 = f1_score(y_valid, y_valid_predicted)
    return [train_acc, test_acc, p, r, f1]

def plot_confusion(conf_matrix_array, ax, title=None):
    X = conf_matrix_array / np.sum(conf_matrix_array, axis=1)[:, None]
    sns.heatmap(X, ax=ax,
                linewidths=1, linecolor='slategrey', cmap='Blues',
                annot=True, xticklabels=['No Buy', 'Buy'], yticklabels=['No Buy', 'Buy'],
                cbar=False, square=True)
    ax.set_title(title)
```

# Data Pre-processing

## Train/Validation/Test Split

```python
df = pd.read_csv("../data/processed/cleaned_df.csv")
print(f" Data shape: {df.shape}")
X = df.drop(columns = ['customer_id', 'buy'])
y = df['buy']
seed = 123
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=seed) # get test set
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=0.4, stratify=y_train, random_state=seed) # get train/valid set
print("==========================")
print("   Stratified Splitting   ")
print("==========================")
print(f"Train shape: {X_train.shape}, % buy: {y_train.sum()/len(y_train)*100:.2f}%")
print(f"Valid shape: {X_valid.shape}, % buy: {y_valid.sum()/len(y_valid)*100:.2f}%")
print(f" Test shape: {X_test.shape}, % buy: {y_test.sum()/len(y_test)*100:.2f}%")
print("")
X.head()
```

## Handling Categorical Features

```python
categorical_features = ['accepts_marketing', 'ordered_month', 'gender', 'free_shipping',
                       'product_type', 'skin_type', 'location', 'fv_site']
categorical_transformer = Pipeline(steps =[
                                        ('imputer', SimpleImputer(strategy = 'constant', fill_value = 'unknown')),
                                        ('onehot', OneHotEncoder(handle_unknown = 'ignore'))
])

preprocessor = ColumnTransformer(transformers = [('cat', categorical_transformer, categorical_features)],
                                 remainder = "passthrough")
```

# Modelling Strategies

## Strategy 1: Balancing the Class Weights

```python
models = {
          'Dummy Classifier' : DummyClassifier(),
          'Decision Tree': DecisionTreeClassifier(class_weight="balanced"),
          'Naive Bayes': MultinomialNB(class_prior=[0.5, 0.5]),
          'Logistic Regression': LogisticRegression(max_iter=200, class_weight="balanced"), 
          'XGBoost' : XGBClassifier(scale_pos_weight=(y_train == False).sum()/(y_train == True).sum()),
          'LGBM': LGBMClassifier(class_weight="balanced")
         }
```

```python
results_bal, fig = run_models(models, preprocessor, X_train, y_train, X_valid, y_valid)
```

```python
results_bal.to_csv("results_balanced.csv")
fig.savefig('results_balanced.png')
```

## Strategy 2: Undersampling

```python
# First we need to undersample the data
rus = RandomUnderSampler(random_state=123)
print(f"Original fractions: No buy = {(len(y_train) - y_train.sum())/len(y_train)*100:.2f}%, Buy = {y_train.sum()/len(y_train)*100:.2f}%")
X_train_res, y_train_res = rus.fit_resample(X_train, y_train)
print(f"Resampled fractions: No buy = {(len(y_train_res) - y_train_res.sum())/len(y_train_res)*100:.2f}%, Buy = {y_train_res.sum()/len(y_train_res)*100:.2f}%")
```

```python
models = {
          'Dummy Classifier' : DummyClassifier(),
          'Decision Tree': DecisionTreeClassifier(),
          'Naive Bayes': MultinomialNB(),
          'Logistic Regression': LogisticRegression(), 
          'XGBoost' : XGBClassifier(),
          'LGBM': LGBMClassifier()
         }
```

```python
results_us, fig = run_models(models, preprocessor, X_train_res, y_train_res, X_valid, y_valid)
```

```python
results_us.to_csv("results_undersampled.csv")
fig.savefig('results_undersampled.png')
```

## Strategy 3: Oversampling

WARNING the next two cells are dirty coding!!!

I'm transforming the month columns to a cyclic feature so I can use SMOTENC, this could definitely be cleaned up!

```python
# I'm going to turn the month column into a numerical feature to facilitate SMOTENC
# We aneed to account for the fact that the feature is cyclic, i.e., month 1 is close to month 12
# We can do this using a cos and sin transform
XX_train = X_train.copy()
XX_train['sin_time'] = (np.sin(2 * np.pi * df.ordered_month / 12) * 100).astype(int) / 100
XX_train['cos_time'] = (np.cos(2 * np.pi * df.ordered_month / 12) * 100).astype(int) / 100
XX_train = XX_train.drop(columns = ['ordered_month'])
# Now oversample (this takes a long time)
os = SMOTENC(categorical_features=np.arange(0,8), random_state=123)
print(f"Original fractions: No buy = {(len(y_train) - y_train.sum())/len(y_train)*100:.2f}%, Buy = {y_train.sum()/len(y_train)*100:.2f}%")
X_train_os, y_train_os = os.fit_resample(XX_train, y_train)
print(f"Resampled fractions: No buy = {(len(y_train_os) - y_train_os.sum())/len(y_train_os)*100:.2f}%, Buy = {y_train_os.sum()/len(y_train_os)*100:.2f}%")
```

```python
# now reverse the mapping of months to continuous features, i.e., map them back to integers
i = np.arange(1, 13)
sin_i = (np.sin(2 * np.pi * i / 12) * 100).astype(int) / 100
cos_i = (np.cos(2 * np.pi * i / 12) * 100).astype(int) / 100
X_train_os['ordered_month'] = 1
for month, sin_month, cos_month in zip(i, sin_i, cos_i):
    X_train_os.iloc[X_train_os.query("sin_time == @sin_month & cos_time == @cos_month").index, 9] = month
X_train_os = X_train_os.drop(columns = ['sin_time', 'cos_time'])
```

```python
models = {
          'Dummy Classifier' : DummyClassifier(),
          'Decision Tree': DecisionTreeClassifier(),
          'Naive Bayes': MultinomialNB(),
          'Logistic Regression': LogisticRegression(), 
          'XGBoost' : XGBClassifier(),
          'LGBM': LGBMClassifier()
         }
```

```python
results_os, fig = run_models(models, preprocessor, X_train_os, y_train_os, X_valid, y_valid)
```

```python
results_os.to_csv("results_oversampled.csv")
fig.savefig('results_oversampled.png')
```
